```python
import datetime
import io
import os
import re
from pathlib import Path
from collections import Counter

import pandas as pd

from multiplex_imaging_pipeline.utils import listfiles
```

```python
old_dir = Path('/diskmnt/Projects/Users/estorrs/omero_old/ManagedRepository/mschan_156/')
fps = sorted(listfiles(old_dir, regex=r'(svs)|(tif)$'))
len(fps)
```

```python
counts = {}
sids = []
for fp in fps:
    sid = fp.split('/')[-1].replace('.svs', '').replace('.tif', '')
    
    if sid not in counts:
        counts[sid] = 0
    counts[sid] += 1
    
    count = counts[sid]
    
    if count > 1:
        sid = f'{sid}_{count}'
        
    sids.append(sid)
    
```

```python
cmds = []
for sid, fp in zip(sids, fps):
    name = re.sub(r'( |\(|\))', r"_", sid)
    fp = re.sub(r'( |\(|\))', r"\\\1", fp)
    cmd = f'omero import {fp} -T Project:name:HTAN_old/Dataset:name:{name}'
    cmds.append(cmd)
```

```python
f = open('/diskmnt/Projects/Users/estorrs/sandbox/upload_htan.sh', 'w')
for cmd in cmds:
    f.write(cmd + '\n')
f.close()
```
