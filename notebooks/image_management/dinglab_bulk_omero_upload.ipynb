```python
import datetime
import io
import os
import re
from pathlib import Path
from collections import Counter

import pandas as pd

from multiplex_imaging_pipeline.utils import listfiles
```

```python
old_dir = Path('/diskmnt/Projects/Users/estorrs/omero_old/ManagedRepository/ahouston_213/')
fps = sorted(listfiles(old_dir, regex=r'/[sS][0-9]+.*.tiff?$'))
len(fps), fps[:2]
```

```python
re.sub(r'^.*/([Ss][0-9]+.*)[_ \-][Ss][Cc][Aa][Nn].*$', r'\1',
       '/diskmnt/Projects/Users/estorrs/omero_old/ManagedRepository/ahouston_213/2023-05/04/18-34-33.508/S17-12666-2_Scan1.qptiff'
)
```

```python
counts = {}
sids = []
for fp in fps:
#     fname = fp.split('/')[-1]
    sid = re.sub(r'^.*/([Ss][0-9]+.*)[_ \-][Ss][Cc][Aa][Nn].*$', r'\1', fp)

    if sid not in counts:
        counts[sid] = 0
    counts[sid] += 1
    
    count = counts[sid]
    
    if count > 1:
        sid = f'{sid}_{count}'
        
    sids.append(sid)
len(sids)
```

```python
sids
```

```python
cmds = []
for sid, fp in zip(sids, fps):
    name = re.sub(r'( |\(|\))', r"_", sid)
    fp = re.sub(r'( |\(|\))', r"\\\1", fp)
    cmd = f'omero import {fp} -T Project:name:Dinglab_old/Dataset:name:{name}'
    cmds.append(cmd)
```

```python
f = open('/diskmnt/Projects/Users/estorrs/sandbox/upload_dinglab.sh', 'w')
for cmd in cmds:
    f.write(cmd + '\n')
f.close()
```
